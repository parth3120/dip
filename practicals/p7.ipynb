```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('Original-cameraman-image.png', cv2.IMREAD_GRAYSCALE)

roberts_x = np.array([[1, 0], [0, -1]])
roberts_y = np.array([[0, 1], [-1, 0]])

roberts_x_image = cv2.filter2D(image, -1, roberts_x).astype(np.float32)
roberts_y_image = cv2.filter2D(image, -1, roberts_y).astype(np.float32)

roberts_combined_image = cv2.magnitude(roberts_x_image, roberts_y_image)

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(roberts_x_image, cmap='gray')
plt.title('Robert Operator X')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(roberts_y_image, cmap='gray')
plt.title('Robert Operator Y')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(roberts_combined_image, cmap='gray')
plt.title('Robert Operator Combined')
plt.axis('off')

plt.show()


```
